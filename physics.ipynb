```python
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
import soundfile as sf
import io

# 소리 파일 읽기
audio_path = '/content/sound/delta.m4a'
audio = AudioSegment.from_file(audio_path, format="m4a")

# 1초에서 5초까지의 데이터만 사용
start_time = 1  # 시작 시간 (초)
end_time = 5    # 종료 시간 (초)
audio = audio[start_time * 1000:end_time * 1000]

# m4a 파일을 WAV로 변환
wav_data = io.BytesIO()
audio.export(wav_data, format="wav")
wav_data.seek(0)

# WAV 파일 읽기
audio_data, sample_rate = sf.read(wav_data)

# time 생성
time = np.arange(0, len(audio_data)) / sample_rate + start_time

# 소리 파일 읽기 (두 번째 파일)
audio_path2 = '/content/sound/delta.m4a'
audio2 = AudioSegment.from_file(audio_path2, format="m4a")

# 1초에서 5초까지의 데이터만 사용
audio2 = audio2[start_time * 1000:end_time * 1000]

# m4a 파일을 WAV로 변환
wav_data2 = io.BytesIO()
audio2.export(wav_data2, format="wav")
wav_data2.seek(0)

# WAV 파일 읽기
audio_data2, sample_rate2 = sf.read(wav_data2)

# time 생성
time2 = np.arange(0, len(audio_data2)) / sample_rate2 + start_time

# 두 파형을 반전하여 더하기
combined_audio_data = audio_data + audio_data2[::-1]

# 시간 도메인에서의 파형 시각화
plt.figure(figsize=(12, 4))
plt.plot(time, audio_data, label='Graph 1')
plt.plot(time2, audio_data2, label='Graph 2')
plt.plot(time, combined_audio_data, label='Combined', linestyle='dashed')
plt.title('Combined Waveform in Time Domain')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
import soundfile as sf
import io

# 소리 파일 읽기
audio_path1 = '/content/sound/alpha.m4a'
audio1 = AudioSegment.from_file(audio_path1, format="m4a")

audio_path2 = '/content/sound/delta.m4a'
audio2 = AudioSegment.from_file(audio_path2, format="m4a")

# 5초까지의 데이터만 사용
audio1 = audio1[:5000]
audio2 = audio2[:5000]

# 두 파형을 더하여 새로운 파형 생성
combined_audio = audio1.overlay(audio2)

# 새로운 파형을 WAV로 변환
wav_data_combined = io.BytesIO()
combined_audio.export(wav_data_combined, format="wav")
wav_data_combined.seek(0)

# WAV 파일 읽기
audio_data_combined, sample_rate_combined = sf.read(wav_data_combined)

# 시간 도메인에서의 파형 시각화
time_combined = np.arange(0, len(audio_data_combined)) / sample_rate_combined
plt.figure(figsize=(12, 4))
plt.plot(time_combined, audio_data_combined)
plt.title('Combined Waveform in Time Domain')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
